```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
```

```
dados_corona = pd.read_csv('https://raw.githubusercontent.com/peixebabel/COVID-19/master/data/casos-br-total.csv')

dados_bd = pd.read_csv('https://raw.githubusercontent.com/wjunior15/COVID-19-Brasil/master/CoronaBD.csv')
dados_bd.head()
```

```
#tratando dados nulo

##Desafio 1:

dados_corona.update(dados_corona['Mortes'].fillna(0))

dados_corona.head()
```

```
dados_corona = dados_corona[['Data', 'Confirmados', 'Mortes']]
```

```
plt.figure(figsize = (16,8))
plt.xticks(rotation = 90)
graf1 = sns.lineplot(x = dados_corona.index, y = 'Confirmados', data = dados_corona, marker = 'o', color = 'black')
graf2 = sns.lineplot(x = dados_corona.index, y = 'Mortes', data = dados_corona, color = 'red', marker = 'o')
plt.title('Dias de infecção X Casos confirmados / Mortes - COVID-19', color = 'red')
plt.ylabel('Casos confirmados / Mortes', color = 'red')
plt.xlabel('Dias de infecção', color = 'red')
plt.legend([graf1, graf2], ['Casos confirmados', 'Mortes confirmadas'])
plt.show()
```

```
plt.figure(figsize = (16,8))
plt.xticks(rotation = 90)
sns.lineplot(x = dados_corona.index, y = 'Confirmados', data = dados_corona, marker = 'o', color = 'black')
sns.lineplot(x = dados_corona.index, y = 'Mortes', data = dados_corona, color = 'red', marker = 'o')
plt.title('Dias de infecção X Casos confirmados / Mortes - COVID-19 (Escala Log.)', color = 'red')
plt.ylabel('Casos confirmados / Mortes', color = 'red')
plt.xlabel('Dias de infecção', color = 'red')
plt.yscale('log') #coloca o eixo y em escala logaritmica

#label_point(dados_corona.sepal_length, dados_corona.sepal_width, dados_corona.species, plt.gca()) 

plt.show()
```

```
plt.figure(figsize = (16,8))
plt.xticks(rotation = 90)
graf1 = sns.lineplot(x = dados_bd.index, y = 'Confirmados', data = dados_bd, marker = 'o', color = 'black')
graf2 = sns.barplot(x = dados_bd.index, y = 'Novos casos', data = dados_bd, color = 'red')
plt.title('Dias de infecção X Casos confirmados / Novos casos - COVID-19 Bom Despacho-MG', color = 'red')
plt.ylabel('Casos confirmados / Novos casos', color = 'red')
plt.xlabel('Dias de infecção', color = 'red')
plt.legend([graf1, graf2], ['Casos confirmados', 'Novos casos'])
plt.show()
```
